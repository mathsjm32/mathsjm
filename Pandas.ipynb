# Pandas
Pandas는 Series(1차원 테이블 데이터용)와 DataFrame(2차원 테이블 데이터용)이라는 강력한 데이터구조를 제공하여 테이블 형태의 데이터를 쉽게 조작하고 분석할 수 있는 필수적인 도구이다.
Pandas를 사용하면 데이터 필터링, 그룹화, 통계 계산 등 다양한 작업을 직관적이고 효율적으로 수행할 수 있다.

Pandas는 구조적 데이터 처리에 강점을 두고 있으며, Numpy는 대규모 수치 연산에서 고속 성능을 제공한다. 따라서, 데이터의 성격과 작업 목적에 따라 두 라이브러리를 함께 사용하는게 일반적이다.

```python
import pandas as pd
x = [10,20,30,40]
y = pd.Series(x)
print(y)
```

```python
x = [10,20,30,40]
y = ['a', 'b', 'c', 'd']
z = pd.Series(x, index=y)
print(z)
```

```python
x = {'apple':1, 'banana':2, 'cherry':3}
y = pd.Series(x)
print(y)
```

```python
x = pd.Series([10,20,30,40,50,60,70])

print(x.dtype)
print(x.index)
print(x.values) # Numpy 배열로 반환

print(x.head()) # 상위 5개 출력
print(x.tail(2)) # 하위 2개 출력 (기본값=5)
```

```python
# 인덱싱 및 슬라이싱 (iloc & loc)
x = pd.Series([10,20,30,40,50,60,70])
# iloc(정수 위치 기반 인덱싱)
print(x.iloc[1])
print(x.iloc[1:4])
# loc(인덱스 레이블 기반 인덱싱)
y = pd.Series([10,20,30,40], index=['a','b','c','d'])
print(y.loc['b'])
print(y.loc['b':'d'])
```

```python
# concat() - 데이터 추가
x = pd.Series([10,20,30], index=['a','b','c'])
y = pd.Series([40,50], index=['d','e'])

z = pd.concat([x,y])
print(z)
```

```python
x = pd.Series([10,20,30], index=['a','b','c'])
x['f']=60
print(x)
```

```python
x = pd.Series([10,20,30,40], index=['a','b','c','d'])
x['b']=25
print(x)
```

```python
x = pd.Series([10,20,30,40], index=['a','b','c','d'])
x[['a','d']] = [15,45]
print(x)
```

```python
x = pd.Series([10,20,30,40,50], index=['a','b','c','d','e'])
x[x>30] = x[x>30] + 10
# x[x>30] += 10
print(x)
```

```python
x = pd.Series([10,20,30], index=['a','b','c'])
x = x.rename({'a':'x', 'b':'y'})
print(x)
```

```python
x = pd.Series([10,20,30,40,50])

print("평균: ", x.mean())
print("합계: ", x.sum())
print("중위수: ", x.median())
print("최댓값: ", x.max())
print("최솟값: ", x.min())
print("분산: ", x.var())
print("표준편차: ", x.std())
```

```python
print(x.describe())
```

```python
x = pd.Series(['a','b','c','b','a'])
print(x.value_counts()) # 빈도수 확인
print(x.value_counts(normalize=True)) # 빈도 비율 확인
```

```python
# 데이터 필터링
x = pd.Series([10,20,30,40,50], index=['a','b','c','d','e'])
print(x[x>25])
print("AND 조건 필터링 결과:")
print(x[(x>25)&(x<35)])
print("OR 조건 필터링 결과:")
print(x[(x<25)|(x>35)])
```

```python
# 1. sort_values() 를 사용한 값 기준 정렬
x = pd.Series([20,10,40,30], index=['a','b','c','d'])
print(x.sort_values()) # 기본 값: 오름차순
print(x.sort_values(ascending=False)) # 기본 값: 내림차순
```

```python
# 2. sort_index() 를 사용한 인덱스 기준 정렬
x = pd.Series([20,10,40,30], index=['a','b','c','d'])
print(x.sort_index()) # 기본 값: 오름차순
print(x.sort_index(ascending=False)) # 기본 값: 내림차순
```

```python
# 데이터 그룹화 및 집계 (groupby & agg)
x = pd.Series([10,20,30,10,20,30], index=['a','b','a','b','a','b'])

print(x.groupby(level=0).sum()) # level=0은 가장 바깥쪽 인덱스를 의미
print(x.groupby(level=0).mean())
print(x.groupby(level=0).agg(['sum','mean','count']))
```

```python
x = {
    'Name':['Ant','Bee','Cat', 'Dog'],
    'Age':[25,30,35,40],
    'City':['Seoul', 'Busan', 'Incheon', 'Daejeon'],
}

df = pd.DataFrame(x, index=['A', 'B', 'C', 'D'])
print(df)
```

```python
# iloc(정수 위치 기반 인덱싱)
print(df.iloc[1,2]) # 1행 2열
print(df.iloc[1:2, 0:2]) # 1행의 0,1열
```

```python
# loc(인덱스 레이블 기반 인덱싱)
print(df.loc['A', 'City']) # A행의 City열
print(df.loc['A':'B', 'Name':'Age']) # A,B행의 Name~Age열
```

```python
df = df.assign(Cost=[10,30,150,150]) # df.assign() - 새로운 열 추가
print(df)
```

```python
# 열 추가 및 수정
df['Needs'] = [5,20,60,63]
df['Age'] = [20,25,30,40]
print(df)
```

```python
df.loc['A', 'Needs'] = 10
print(df)
```

```python
df.iloc[1,3] = 50
print(df)
```

```python
df.loc[['A','C'], 'Cost'] = 20,100
print(df)
```

```python
df.loc[df['Cost']>=100, 'Passed'] = True
print(df)
```

```python
df.loc[(df['Age']>20) & (df['Name']=='Bee'), 'Needs'] = 40
print(df)
```

```python
df = df.rename(columns={'Name':'Full Name', 'City':'Location'})
print(df)
```

```python
df = pd.DataFrame({
    'Math':[85,90,95,100,80],
    'Science':[88,92,85,89,94]
})
print(df.describe())
```

```python
math_mean = df['Math'].mean()
science_sum = df['Science'].sum()

print(f"Math 평균: {math_mean}")
print("Science 합계: ", science_sum)
```

```python
# 빈도 수 확인
import pandas as pd
df = pd.DataFrame({
    'Fruit':['apple','banana','apple','orange','banana','apple'],
    'Color':['red','yellow','red','orange','yellow','red']
})

print(df['Fruit'].value_counts())
print(df.apply(pd.Series.value_counts))
```

```python
for col in df.columns:
    print(f"'{col}' 열의 고윳값 빈도: ")
    print(df[col].value_counts(), '\n')
```

```python
df1 = pd.DataFrame({
    'ID':[1,2,3],
    'Name':['Ant','Bee','Cat']
})
df2 = pd.DataFrame({
    'ID':[2,3,4],
    'Score':[88,92,85]
})
display(df1, df2)
```

```python
# 데이터 병합 및 결합
# merge(df1, df2, on='target col', how='결합 방식')
# 결합방식: inner, outer, left, right
merged_inner = pd.merge(df1,df2,on='ID',how='inner')
merged_left = pd.merge(df1,df2,on='ID',how='left')
print(merged_inner)
print(merged_left)
```

```python
# 데이터 결합 - pd.concat()
# axis=0 : 행 방향, axis=1 : 열 방향
print(pd.concat([df1,df2], axis=0))
print(pd.concat([df1,df2], axis=1))
```

```python
df = pd.DataFrame({
    'Name':['Ant','Bee','Cat','Dog'],
    'Age':[24,27,22,32],
    'Score':[88,95,85,90]
})
# 데이터 정렬
df.sort_values(by='Age') # Age 기준 오름차순 정렬
df.sort_values(by='Score', ascending=False) # Score 기준 내림차순 정렬
df.sort_values(by=['Age', 'Score'], ascending=[True, False]) # Age 기준 오름차순, Score 내림차순 정렬
df.sort_index() # 행 인덱스 기준 오름차순 정렬
df.sort_index(axis=1, ascending=False) # 열 인덱스 기준 내림차순 정렬
```

```python
df = pd.DataFrame({
    'Category':['A','A','B','B','A','B'],
    'Type':['X','Y','X','Y','X','Y'],
    'Values':[10,20,30,40,50,60]
})

# 데이터 그룹화 및 집계
print(df.groupby('Category')['Values'].sum())
print(df.groupby(['Category','Type'])['Values'].mean())
print(df.groupby('Category')['Values'].agg(['sum','mean','count']))
```
